```python
#برای داده های پرت از ای کیو ار استفاده میکنیم
#نباید منفی باشه پس دیتاهای بزرگتر از صفر رو میخوایم
data = data[data['minimum nights']>0]
Q1 = data ['minimum nights'].quantile(0.25)
Q3 = data ['minimum nights'].quantile(0.75)
IQR = Q3-Q1
lower = Q1-1.5*IQR
upper = Q3+1.5*IQR
data = data [(data ['minimum nights']>=lower)&(data ['minimum nights']<=upper)]
```

```python
data.describe()
```

```python
#حذف با z_score
#بر اساس میانگین و انحراف معیار
#هر کدام از داده ها منهای میانگین تقسیم بر انحراف معیار
#دو تا روش رو هم مطالعه کن خوبه: lof & isolation forest & robust z_score
z = (data ['price']-data['price'].mean())/data['price'].std()
data = data [(z>-3) & (z<3)]
z = (data ['service fee']-data['service fee'].mean())/data['service fee'].std()
data = data [(z>-3) & (z<3)]
```
